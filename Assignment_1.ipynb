You have a matrix $X$ of size (10,5) filled with random values. Print the following :
* Minimum value and index where it occured, for each column
* Maximum value and the index where it occured, for each column
* Minimum value and index where it occured, for each row
* Maximum value and index where it occured, for each row
* Minimum value and index where it occured, for the entire matrix
* Maximum value and index where it occured, for the entire matrix

````python
np.random.seed(0)
X = np.random.rand(10,5)

minInCols = np.amin(X, axis=0)
MinColsInd = np.argmin(X, axis=0)
print("Minimum values per column :", minInCols)
print("Indices of minimum values per column :",MinColsInd)

maxInCols = np.amax(X, axis=0)
MaxColsInd = np.argmax(X, axis=0)
print("Maximum values per column :", maxInCols)
print("Indices of maxmum values per column :",MaxColsInd)

minInRows = np.amin(X, axis=1)
MinRowsInd = np.argmin(X, axis=1)
print("Minimum values per Rows :", minInRows)
print("Indices of minimum values per Rows :",MinRowsInd)

maxInRows = np.amax(X, axis=1)
MaxRowsInd = np.argmax(X, axis=1)
print("Maximum values per Rows :", maxInRows)
print("Indices of maxmum values per Rows :",MaxRowsInd)

min_X = np.amin(minInCols, axis=0)
MinInd = np.argmin(minInCols, axis=0)
cordmin = (MinColsInd[MinInd],MinInd)
print("Minimum value of matrix :", min_X)
print("Indices of minimum value of matrix  :",cordmin)

max_X = np.amax(maxInCols, axis=0)
MaxInd = np.argmax(maxInCols, axis=0)
cordmax = (MaxColsInd[MaxInd],MaxInd)
print("Minimum value of matrix :", max_X)
print("Indices of minimum value of matrix :",cordmax)
````

Let $Y$ denote a square matrix of size 10 with random values. Print the following:
* Determinant
* Rank of the matrix
* Diagonal Elements
* Trace
* Eigen values
* Matrices obtained after Singular Value Decomposition
* Inverse if exists, otherwise Pseudo-Inverse
* Multiplication $Y$ with $X$
* Element-wise multiplication of $Y$ and its inverse(or pseudo-inverse)

````python
np.random.seed(3)
Y = np.random.rand(10,10)

det = np.linalg.det(Y)
print('Deterinmant of the matrix Y is',det)
rank = np.linalg.matrix_rank(Y)
print('Rank of the matrix Y is', rank)

dia = np.diagonal(Y)
print('Diagonal elements of Y is',dia)

trace = np.sum(dia)
print('Trace of the matrix Y is ',trace)

val, vec = np.linalg.eig(Y)
print('Eigen values of the matrix Y is',val)

u, s, vh = np.linalg.svd(Y, full_matrices=True)
print("Matrices obtained after Singular Value Decomposition are:")
print(u)
print(s)
print(vh)
````

Plot the sine function with domain in $[-3\pi,3\pi]$. Label the axes appropriately. Add title $\sin(x)$. Use latex formatting while adding the title. Set the plot size to (10,10)

````python
import math 
v = 3*math.pi
time = np.arange(-v, v, 0.1);
amplitude   = np.sin(time)
fig= plt.figure(figsize=(10,10))
plt.plot(time, amplitude)
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.title('sine wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
````

Plot $\sin(x)$ and its spectrum, $\cos(x)$ and its spectrum in a single plot, in the domain $[-2\pi,2\pi]$. Label the plot appropriately.

````python
v = 2*math.pi
time = np.arange(-v, v, 0.1);
amplitude   = np.sin(time)

fig= plt.figure(figsize=(20,20))

plt.subplot(2, 2, 1)
plt.plot(time, amplitude)
plt.grid(True, which='both')
plt.title('sine wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.axhline(y=0, color='k')

plt.subplot(2, 2, 2)
#plt.plot(x, y)
sp = np.fft.fft(amplitude)
freq = np.fft.fftfreq(time.shape[-1])
plt.plot(freq, sp.real, freq, sp.imag)
plt.title('sin fourier')
plt.xlabel('frequency')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.show()

plt.show()
````

````python
v = 2*math.pi
time = np.arange(-v, v, 0.1);
amplitude   = np.cos(time)

fig= plt.figure(figsize=(20,20))

plt.subplot(2, 2, 1)
plt.plot(time, amplitude)
plt.title('cos wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')

plt.subplot(2, 2, 2)
sp = np.fft.fft(amplitude)
freq = np.fft.fftfreq(time.shape[-1])
plt.plot(freq, sp.real, freq, sp.imag)
plt.title('cos fourier')
plt.xlabel('frequency')
plt.ylabel('Amplitude')
plt.grid(True, which='both')

plt.show()
````
